```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Make a reference to the books.csv file path
file_name=os.path.join("Resources","books.csv")
# Import the books.csv file as a DataFrame with utf-8 encoding.
file_df=pd.read_csv(file_name,encoding="utf-8")
file_df.head()
```

```python
#  Get the data types of each column.
file_df.dtypes
```

```python
# Get a list of all columns within the DataFrame
column_list=file_df.columns 
```

```python
# Remove unecessary columns from the DataFrame and save the new DataFrame
# Only keep: "isbn", "original_publication_year", "original_title", "authors",
# "ratings_1", "ratings_2", "ratings_3", "ratings_4", "ratings_5"
keep_columns=["isbn", "original_publication_year", "original_title", "authors","ratings_1", "ratings_2", "ratings_3", "ratings_4", "ratings_5"]
drop_columns=[]
for df_columns in column_list:
    #print(f"All column names is {df_columns}")
    if df_columns not in keep_columns:
        #print(f"Drop column names is {df_columns}")
        drop_columns.append(df_columns)#drop_columns
#print(drop_columns)
file_df2 = file_df.drop(columns=drop_columns)
file_df2.head()
#file_df.head()
```

```python
# Rename the headers to be more explanatory
file_df2.rename(columns={"isbn":"ISBN","authors":"Author Name","original_publication_year":"Original Publication Year","original_title":"Original Title","ratings_1":"Ratings 1","ratings_2":"Ratings 2","ratings_3":"Ratings 3","ratings_4":"Ratings 4","ratings_5":"Ratings 5"})
 
```
